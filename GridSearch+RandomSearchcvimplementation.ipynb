```python
import pandas as pd
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
from sklearn.model_selection import RandomizedSearchCV
```

```python
# C:\Users\dell\Downloads\student-lifestyle-and-stress-dataset.csv
data = pd.read_csv("student-lifestyle-and-stress-dataset.csv")
y = data['Stress_Level']
```

```python
#D:\c programs\machine_learning_algorithms\cleaner.pkl
cleaner = joblib.load('cleaner.pkl')
```

```python
cleaned_data = cleaner.fit_transform(data)
```

```python
columns = cleaner.get_feature_names_out()
```

```python
print(cleaned_data)
print(columns)
```

```python
modified_data = pd.DataFrame(cleaned_data, columns=columns)
```

```python
print(modified_data)
```

```python
model_rf = RandomForestClassifier(random_state=42)
model_params = {
    'n_estimators' : [50, 100, 150, 200, 300, 350, 400, 450],
'criterion' : ['gini', 'entropy'],
    'max_features' : ['sqrt', 'log2']
}
```

```python
grid = GridSearchCV(estimator=model_rf, param_grid=model_params, cv=5, n_jobs=-1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(modified_data,y, test_size=0.3,  random_state=42)
```

```python
grid.fit(X_train, Y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
print(accuracy_score(Y_test, y_pred))
```

```python
print(classification_report(Y_test, y_pred))
```

```python
print(f'Best parameters : {grid.best_params_}')
```

```python
print(f'Best Score : {grid.best_score_}')
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(Y_test, y_pred, color='blue', linewidth=2)
```

RandomSearch CV


```python
rs = RandomizedSearchCV(estimator=RandomForestClassifier(), param_distributions=model_params, cv=5, n_iter=30, n_jobs=-1, random_state=42)

rs.fit(X_train, Y_train)
```

```python
y_pred = rs.predict(X_test)
```

```python
print(accuracy_score(Y_test, y_pred)*100)
```
